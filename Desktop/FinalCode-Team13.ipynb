```python
def kl_div(m1, s1, m2, s2):
    return np.log(s2 / s1) + (s1 ** 2 + (m1 - m2)**2) / (2 * s2 ** 2) - 0.5

def sym_kl_div(m1, s1, m2, s2):
    return 0.5 * kl_div(m1, s1, m2, s2) + 0.5 * kl_div(m2, s2, m1, s1)

kl_div_rank = {}
for gene_col, gene_id in enumerate(gene_lst):
    for organ_row, organ_name in enumerate(organ_lst):
        if tcga_tables[2, organ_row, gene_col] < min_data_cnt or gtex_tables[2, organ_row, gene_col] < min_data_cnt:
            continue
        kl_div_rank[(organ_name, gene_id)] = sym_kl_div(tcga_tables[0, organ_row, gene_col], 
                                                       tcga_tables[1, organ_row, gene_col] + 1, 
                                                       gtex_tables[0, organ_row, gene_col], 
                                                       gtex_tables[1, organ_row, gene_col] + 1)
ranked_kl_pair = sorted(kl_div_rank.items(), key=lambda x: -x[1])
print ranked_kl_pair[:20]
```
